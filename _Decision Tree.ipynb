## Decision Tree Classifier Implementation with POST Pruning

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
dataset=load_iris()
```

```python
print(dataset.DESCR)
#for description of the data 
```

```python
import seaborn as sns 
df=sns.load_dataset("iris")
```

```python
#Independent and dependent features
x=df.iloc[:,:-1]
y=dataset.target
```

```python
x
```

```python
y
```

```python
len(y),len(x)
#Both have same no. of instances
```

```python
## Train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(
x,y,test_size=0.33,random_state=42)
```

## Major INfo No need to perform standardization

```python
## Decision tree implimenatation
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
##Post prunning

treeclassifier=DecisionTreeClassifier()
```

```python
treeclassifier.fit(x_train,y_train)
```

```python
x_train.head()
```

```python
#tree represenstation
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treeclassifier,filled=True)
```

```python
##Post Prunning with max_depth = 2
treeclassifier=DecisionTreeClassifier(max_depth=2)
treeclassifier.fit(x_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treeclassifier,filled=True)
```

```python
## prediction
y_pred=treeclassifier.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score=accuracy_score(y_pred,y_test)
print(score)
print(classification_report(y_pred,y_test))
```

## DecisionTree Preprunning And Hyperparameter Tunning

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
parameter={
    "criterion":["gini","entropy","log_loss"],
    "splitter":["best","random"],
    "max_depth":[1,2,3,4,5],
    "max_features":["auto","sqrt","log2"]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
clf=GridSearchCV(DecisionTreeClassifier(),param_grid=parameter,cv=5,scoring="accuracy")
```

```python
## train the data
clf.fit(x_train,y_train)
```

```python
clf.best_params_ # Pre Prunning Parameters
```

```python
y_pred=clf.predict(x_test)
```

```python
score=accuracy_score(y_pred,y_test)
```

```python
score
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score=accuracy_score(y_pred,y_test)
print(score)
print(classification_report(y_pred,y_test))
```
